####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig = plt.figure()
```

```python
ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,2)
ax3 = fig.add_subplot(2,3,3)
ax4 = fig.add_subplot(2,3,4)
ax5 = fig.add_subplot(2,3,5)
```

```python
# Cria o gráfico de barras
ax1.bar(sinasc['APGAR1'].value_counts().index, sinasc['APGAR1'].value_counts().values, data= sinasc['APGAR1'])
ax2.bar(sinasc['APGAR5'].value_counts().index, sinasc['APGAR5'].value_counts().values, data= sinasc['APGAR5'])
ax3.bar(sinasc['SEXO'].value_counts().index, sinasc['SEXO'].value_counts().values, data= sinasc['SEXO'])
ax4.bar(sinasc['GESTACAO'].value_counts().index, sinasc['GESTACAO'].value_counts().values, data= sinasc['GESTACAO'])
ax5.bar(sinasc['ESTCIVMAE'].value_counts().index, sinasc['ESTCIVMAE'].value_counts().values, data= sinasc['ESTCIVMAE'])

# Adiciona títulos aos subplots
ax1.set_title('APGAR 1 minuto')
ax2.set_title('APGAR 5 minutos')
ax3.set_title('Sexo')
ax4.set_title('Gestação (semanas)')
ax5.set_title('Estado civil da mãe')

# Adiciona rótulos aos eixos
ax1.set_xlabel('Nascimento')
ax1.set_ylabel('Valor APGAR')
ax2.set_xlabel('Nascimento')
ax2.set_ylabel('Valor APGAR')
ax3.set_xlabel('Sexo')
ax3.set_ylabel('Frequência')
ax4.set_xlabel('Nascimento')
ax4.set_ylabel('Gestação (semanas)')
ax5.set_xlabel('Estado civil da mãe')
ax5.set_ylabel('Frequência')

#rotacionar as legendas do eixo x
ax3.tick_params(axis = 'x' , labelrotation = 90)
ax4.tick_params(axis = 'x' , labelrotation = 90)
# Ajusta o tamanho da figura
plt.subplots_adjust(hspace=1.5)
plt.subplots_adjust(wspace=1.5)
# Mostra a figura
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure', figsize=(6, 4))
sns.barplot( y='CONSULTAS', x='APGAR5', data=sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.3, hspace=0.5)
sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc)


# Rotaciona os rótulos do eixo x em 30 graus
for label in axes[0, 1].get_xticklabels():
    label.set_rotation(30)
```
